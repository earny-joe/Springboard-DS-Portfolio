```python
data = r2['data']
type(data)
```

```python
matrix = [[col for col in row] for row in data]

open_price_list = []
for row in matrix:
    open_price_list.append(row[1])
    
max(filter(lambda x: x is not None, open_price_list))
```

```python
min(filter(lambda x: x is not None, open_price_list))
```

## 4. What was the largest change in any one day (based on High and Low price)?

```python
data[0]
```

```python
high_price_list = []

for row in matrix:
    high_price_list.append(row[2])
    
high_price_list
```

```python
low_price_list = []

for row in matrix:
    low_price_list.append(row[3])
    
low_price_list
```

```python
zipped_list = list(zip(high_price_list, low_price_list))
zipped_list
```

```python
list_of_tuples = zipped_list
high_low_list = [list(elem) for elem in list_of_tuples]

high_low_list
```

```python
dayDiff = [high - low for high, low in high_low_list]
dayDiff
```

```python
max(dayDiff)
```

## 5. What was the largest change between any two days (based on Closing Price)?

```python
close_price_list = []

for row in matrix:
    close_price_list.append(row[4])
    
close_price_list
```

```python
for counter, value in enumerate(close_price_list):
    print(counter, value)
```

```python
test_diff = [abs(close_price_list[i+1] - close_price_list[i]) for i in range(len(close_price_list) - 1)]
max(test_diff)
```

## (Optional) 7. What was the median trading volume during this year. (Note: you may need to implement your own function for calculating the median.)

```python
def middle_num(lst):
    srt = sorted(lst)
    mid = len(lst)//2
    if len(lst) % 2:
        return srt[mid]
    else:
        med = (srt[mid] + srt[mid-1]) / 2.0
        return med
```

```python
middle_num(traded_volume)
```
