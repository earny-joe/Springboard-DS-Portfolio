## Impact of Analytics on the NBA

While Charles Barkley is one of the greatest to play the game of basketball, it cannot be denied that the use of analytics has become widespread in the NBA and is having a significant impact on how it is played. 

Essentially what analytics has done is to shift how decisions are made in the NBA, which relied upon the 'intuition' of basketball experts (i.e. scouts) to a model that now primarily focuses on the analysis of data. 

Why is this so significant? Well, it upended the NBA's status quo. Scouts and their respective opinions on players had played a key role in determining a team's lineup. While these 'experts' may know the ins and outs of basketball, they are still human and susceptible to the biases of the human mind. 

In the book _The Undoing Project_ by Michael Lewis Lewis talks about Daryl Morey, the General Manager of the Houston Rockets (and reason for Charles Barkley's tirade), in the first chapter of the book and highlights how Morey used models to shape their roster and perhaps more importantly, overcome the susceptability of the human mind to certain biases.

(Side note: Lewis also wrote _Moneyball: The Art of Winning an Unfair Game_, which describes how the Oakland A's used analytics to beat larger market teams with overlooked players)

The confirmation bias is one such example. This bias highlights our tendency to search for, interpret, favor, and recall information in a way that confirms one's preexisting beliefs or hypotheses. [(1)](https://en.wikipedia.org/wiki/Confirmation_bias) Where does this come into play? Well, when scouts are judging an athlete, they generally perform what is known as the 'eye-test', which is largely based on physical attributes. A player will tend to pass an eye-test if they resemble an already successful player. 

However, two players -- Steph Curry and Marc Gasol -- are highlighted in the chapter to show how this test is highly suspect. Gasol was pudgy and had a baby-face, which didn't fit the stereotypical mold of an NBA all-star (which he's done three times now). Additionally, most scouts wouldn't take notice of a short, skinny guard whose defining characteristic was long-range shooting yet Steph Curry is now one of the best players in the league. 

Surely, analytics should not be viewed as the end all be all when it comes to judging talent in the NBA. Yet, it can provide us with insights into such things as players that have been overlooked because they don't fit the traditional mold, or how best exploit weaknesses in a team's defense. While nothing is 100% certain, analytics can take a team from run-of-the-mill to perenial playoff, even championship, contenders. Just look at Morey's Rockets, who've not once had a losing record during his tenure and have one of the highest win totals in the NBA over that period. 

So maybe Charles should stick to working on his golf swing...

#### Observation and Explanation of SOS

SOS stands for strength of schedule. According to [Basketball Reference](https://www.basketball-reference.com/about/glossary.html), this score 'is denominated in points above/below average, where zero is average'. Unfortunately, this sounds somewhat like jargon, so lets take a deeper dive.

Basketball Reference uses SRS (Simple Rating System) figures for SOS. This rating takes into account average point differential and strength of schedule. An example from this [blog post](https://www.sports-reference.com/blog/2015/03/srs-calculation-details/) by Mike Lynch at Sports Reference sums it up nicely: 

    "For instance, the 2006-07 Spurs won games by an average of 8.43 points per game and played a schedule with opponents that were 0.08 points worse than average, giving them an SRS of 8.35. This means they were 8.35 points better than an average team. An average team would have an SRS of 0.0."
    
Here are a few quick notes about the various merits and weaknesses of SRS:

- The number it spits out are easy to interpret -- if Team A's rating is 3 bigger than Team B's, this means that the system thinks Team A is 3 points better than Team B.
- It is a predictive system rather than a retrodictive -- this is a very important distinction. You can use these ratings to answer the question: which team is stronger? i.e. which team is more likely to win a game tomorrow?
- It weights all games equally -- meaning that a meaningless game at the end of the season is given the same weight as all the rest of the games
- It weights all points equally, and therefore ignores wins and losses
- It is easily impressed by blowout victories
- Slightly biased toward offensive-minded teams

### Part I: Offensive, Defensive and Net Rating

```python
# scatter plot of Margin of Victory and Wins
f, ax = plt.subplots(figsize=(10, 6))
ax = sns.scatterplot(x="ORtg", y="W", data=df)
```

#### Observation

It looks like offensive rating and wins have a fairly strong positive relationship, i.e. the higher a team's ORtg the more wins they are likely to have. 

While this helps us understand the relationship, why don't we take it a step further and see if there might be 'cut-off' for teams that make the playoffs and teams that don't!

```python
# scatter plot of ORtg and Wins
f, ax = plt.subplots(figsize=(10, 6))
ax = sns.scatterplot(x="ORtg", y="W", data=df, hue='playoff', style='playoff')
```

#### Observation

That's better. As we can see that while there is a little overlap, there is roughly two groups that is divided around 41 wins. As we can see though there are a few teams that despite having a high ORtg, they did not make the playoffs!

```python
# scatter plot of DRtg and Wins
f, ax = plt.subplots(figsize=(10, 6))
ax = sns.scatterplot(x="DRtg", y="W", data=df, hue='playoff', style='playoff')
```

```python
# scatter plot of NRtg and Wins
f, ax = plt.subplots(figsize=(10, 6))
ax = sns.scatterplot(x="NRtg", y="W", data=df, hue='playoff', style='playoff')
```

#### Observation

Above we have the scatter plots for both defensive rating and net rating and their respective relationships with wins. For the defensive rating, we got a loosely negative relationship that can be generally stated as a higher DRtg equals fewer wins (and thus a lower probability of making the playoffs) which is not surprising. 

Net rating though appears to have the strongest relationship out of the three. Just as a reference net rating takes both offensive rating and defensive rating and provides an estimate of point differential per 100 possessions. 

Why do we care about possessions and not just totals? Well [Sports Reference](https://www.sports-reference.com/blog/2014/06/introducing-per-100-possessions-statistics/) has the perfect explanation: "Per-Possession statistics are an easy way to nullify the distorting effects of pace and compare players more easily across teams and eras."

Teams have different styles, some play slower and run their offense while others play faster, relying more heavily on fast-break scoring opportunities. Obviously a team that plays faster is going to have more possessions per game, however when we focus on statistics from a possession viewpoint, we are able to really see how efficient teams are with their possessions, no matter their playing style.
